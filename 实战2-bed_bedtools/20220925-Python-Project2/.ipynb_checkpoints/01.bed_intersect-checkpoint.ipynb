```python
# step1 load reference order

ref_order_dict = {}

ref_fai = open("./hg38_only_chromosome.fa.fai", "r")

index = 0 
for line in ref_fai:
    line_list = line.strip().split("\t")
    ref_order_dict[line_list[0]] = index
    index += 1 
    
ref_fai.close()
        
ref_order_dict
```

```python
# step2 test interect

# 2.1 open file 
bed_a = open("./CTCF_rep1.sort.bed", "r")
bed_b = open("./CTCF_rep2.sort.bed", "r")

# 2.2 init
line_a = bed_a.readline()
line_b = bed_b.readline()

overlap_count = 0 

while line_a and line_b:
    
    line_list_a = line_a.strip().split("\t")
    line_list_b = line_b.strip().split("\t")
    
    # same chromosome
    if line_list_a[0] == line_list_b[0]:
        start_a = int(line_list_a[1])
        end_a = int(line_list_a[2])
        
        start_b = int(line_list_b[1])
        end_b = int(line_list_b[2])        
        
        # no-onverlap a upstream
        if end_a < start_b:
            line_a = bed_a.readline()
        
        # no-overlap a downstream
        elif start_a > end_b:
            line_b = bed_b.readline()
        
        # overlap !!!
        else:
            overlap_count += 1
            temp_line_a = line_a
            temp_line_b = line_b
            line_a = bed_a.readline()
            line_b = bed_b.readline()
    
    # differ chromosome
    else:
        order_a = ref_order_dict.get(line_list_a[0])
        order_b = ref_order_dict.get(line_list_b[0])
        
        if order_a < order_b:
            line_a = bed_a.readline()
        
        else:
            line_b = bed_b.readline()
    

print(overlap_count)
```
